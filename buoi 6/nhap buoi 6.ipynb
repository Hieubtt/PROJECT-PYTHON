```python
class People():
    def __init__(self,name,age,height,weight):
        self.name = name
        self.age = age 
        self.height = height
        self.weight = weight
    
    def info_people(self):
        return(f'Tôi tên là {self.name} , tôi {self.age} tuổi , chiều cao {int(self.height)/10} , nặng {self.weight}')
    
    def get_BMI (self):
        return self.weight / (self.height** 2)
    
    def your_dream (self,kg):
        return f'Cân nặng của tôi hiện tại là {self.weight} kg và tôi muốn giảm về {kg} kg'
    
    

people_1 = People('Hiếu',24,178,86)
print(f'{people_1.info_people()}')
print(f'Số BMI của tôi là : {people_1.get_BMI()}')

print(people_1.your_dream(70))
```
